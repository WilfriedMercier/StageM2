# Plotting data

## Generating the grid

```python
X, Y = np.meshgrid(np.arange(xLen), np.arange(yLen))
```

## Flux plot

```python
plt.rcParams["figure.figsize"] = (11, 11) # (w, h)
f = plt.figure()

ax1 = plt.subplot(111)
ax1.yaxis.set_ticks_position('both')
ax1.xaxis.set_ticks_position('both')
ax1.tick_params(which='both', direction='in', labelsize=26)
ax1.axes.get_xaxis().set_ticklabels([])
ax1.axes.get_yaxis().set_ticklabels([])

im  = plt.imshow(data, cmap='gray_r', origin='bottom')
col = plt.colorbar(im, fraction=0.045)
col.ax.tick_params(labelsize=24)
col.set_label(r'${\rm{Flux}} \,\, [10^{-20} \,\, \rm{erg \, s ^{-1} \, cm^{-2} \, \AA^{-1}}]$', size=24)

plt.savefig('graphics/CGr30bs_139_flux.pdf', bbox_inches='tight')
```

## Velocity field plot

```python
plt.rcParams["figure.figsize"] = (11, 11) # (w, h)
f = plt.figure()

ax1 = plt.subplot(111)
ax1.yaxis.set_ticks_position('both')
ax1.xaxis.set_ticks_position('both')
ax1.tick_params(which='both', direction='in', labelsize=26)
ax1.axes.get_xaxis().set_ticklabels([])
ax1.axes.get_yaxis().set_ticklabels([])

im  = plt.imshow(data, cmap='seismic', origin='bottom', vmin=-85, vmax=85)
col = plt.colorbar(im, fraction=0.045)
col.ax.tick_params(labelsize=24)
col.set_label(r'$\rm{Velocity \,\, [km \, s^{-1}]}$', size=24)

plt.savefig('graphics/CGr30bs_139_vel.pdf', bbox_inches='tight')
```

## Velocity dispersion

```python
plt.rcParams["figure.figsize"] = (11, 11) # (w, h)
f = plt.figure()

ax1 = plt.subplot(111)
ax1.yaxis.set_ticks_position('both')
ax1.xaxis.set_ticks_position('both')
ax1.tick_params(which='both', direction='in', labelsize=26)
ax1.axes.get_xaxis().set_ticklabels([])
ax1.axes.get_yaxis().set_ticklabels([])

im  = plt.imshow(data, cmap='hot_r', origin='bottom', vmin=29, vmax=85)
col = plt.colorbar(im, fraction=0.045)
col.ax.tick_params(labelsize=24)
col.set_label(r'$\rm{Velocity \,\, dispersion \,\, [km \, s^{-1}]}$', size=24)

plt.savefig('graphics/CGr30bs_139_disp.pdf', bbox_inches='tight')
```
